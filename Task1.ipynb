```python
# Split the single column into multiple columns using ';' as the delimiter
df_cleaned = df['school;sex;age;address;famsize;Pstatus;Medu;Fedu;Mjob;Fjob;reason;guardian;traveltime;studytime;failures;schoolsup;famsup;paid;activities;nursery;higher;internet;romantic;famrel;freetime;goout;Dalc;Walc;health;absences;G1;G2;G3']\
    .str.split(';', expand=True)

# Display the cleaned data info
df_cleaned.info(), df_cleaned.head()
```

```python
# Assign proper column names
column_names = [
    'school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob',
    'reason', 'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid',
    'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout',
    'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2', 'G3'
]
df_cleaned.columns = column_names

# Remove quotes and convert to appropriate types
df_cleaned = df_cleaned.replace('"', '', regex=True)
df_cleaned['age'] = pd.to_numeric(df_cleaned['age'], errors='coerce')
df_cleaned['G1'] = pd.to_numeric(df_cleaned['G1'], errors='coerce')
df_cleaned['G2'] = pd.to_numeric(df_cleaned['G2'], errors='coerce')
df_cleaned['G3'] = pd.to_numeric(df_cleaned['G3'], errors='coerce')

# Sort by 'age' and 'G3' for cleaner analysis
df_cleaned_sorted = df_cleaned.sort_values(by=['age', 'G3'], ascending=[True, False])

# Display cleaned and sorted data
df_cleaned_sorted.head(10)
```

## 1. What is the average score in math (G3)?

```python
avg_math_score = df_cleaned_sorted['G3'].mean()
print(f"Average Score in Math (G3): {avg_math_score:.2f}")
```

- The : .2f format specifier is used to round the value to 2 decimal places and format it as a floating-point number.  
- Why use `mean()`? 
- It helps to find the central tendency of a dataset.  
- In your case, `mean()` is used to calculate the average score of students in the `G3` column (final grade).

# 3. Create a bar chart comparing the average scores of male and female students.


- A bar chart is useful for comparing the mean values of groups (male vs. female).
- It shows the difference in average scores between male and female students.

```python

df_melted = df_cleaned_sorted.melt(id_vars=['sex'], value_vars=['G1', 'G2', 'G3'], var_name='Grade_Type', value_name='Score')

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y='Score', hue='Grade_Type', data=df_melted, ci=None, palette='viridis')
plt.title('Average Scores of G1, G2, and G3 by Gender')
plt.xlabel('Gender (M/F)')
plt.ylabel('Average Score')
plt.show()
```

- **df.melt()**
   - Converts the DataFrame from wide format to long format â€” necessary for plotting multiple values side by side.
- **HUE**
   - hue is used to group data based on a categorical variable and plot it using different colors.
- **CI**
   - ci stands for Confidence Interval
   - it shows the range of uncertainty for the mean value.
   - Setting ci=None removes the confidence interval.
