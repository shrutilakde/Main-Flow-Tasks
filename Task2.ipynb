Detect and handle outliers using statistical techniques (e.g., IQR or
Z-scores).
see here are lot of outliers so we have remove all this outliers

```python

import numpy as np

# Load dataset


# Select only numerical columns
num_cols = ['Discount', 'Profit', 'Quantity', 'Sales', 'Shipping.Cost']

# Apply IQR method to remove outliers
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Save cleaned dataset
df.to_csv("cleaned_dataset.csv", index=False)

print("Outliers removed successfully!")
```

```python
# Function to remove outliers using IQR method
def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)  # First quartile (25th percentile)
        Q3 = df[col].quantile(0.75)  # Third quartile (75th percentile)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]  # Filter data
    return df

# Apply outlier removal
filtered_df = remove_outliers_iqr(df, ["Profit", "Sales", "Quantity", "Shipping.Cost"])

# Display new dataset size after outlier removal
df.shape, filtered_df.shape
```

```python
# Plot boxplots before and after outlier removal for comparison
fig, axes = plt.subplots(2, 4, figsize=(16, 8))

columns = ["Profit", "Sales", "Quantity", "Shipping.Cost"]

for i, col in enumerate(columns):
    # Boxplot before outlier removal
    sns.boxplot(y=df[col], color="lightcoral", ax=axes[0, i])
    axes[0, i].set_title(f"Before Outlier Removal: {col}")

    # Boxplot after outlier removal
    sns.boxplot(y=filtered_df[col], color="lightblue", ax=axes[1, i])
    axes[1, i].set_title(f"After Outlier Removal: {col}")

plt.tight_layout()
plt.show()
```
